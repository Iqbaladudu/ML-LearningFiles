## Import Matplotlib & Seaborn

```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

## Load Data

```python
df_iris = pd.read_csv('dataset/iris.data', names=[
    'sepal_length',
    'sepal_width',
    'petal_length',
    'petal_width',
    'Class'
])
```

# Data Visualization

## Histogram - Distribusi Data

```python
df_iris['class'].hist()
plt.show()
```

## Scatter Plot - Distribusi dan Relasi Data

```python
df_iris.plot(kind='scatter', x='sepal_length', y='sepal_width', )
plt.show()
```

```python
df_iris.plot(kind='scatter', x='petal_length', y='petal_width')
plt.show()
```

```python
plt.scatter(df_iris['petal_length'], df_iris['petal_width'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

## Displot - Distribusi Data

```python
sns.displot(df_iris['class'])
plt.show()
```

```python
sns.displot(df_iris['petal_width'])
plt.show()
```

```python
sns.displot(df_iris['sepal_length'], kind='kde')
plt.show()
```

```python
sns.displot(df_iris['petal_length'], kind='ecdf')
plt.show()
```

```python
sns.displot(df_iris['petal_length'], kind='hist')
plt.show()
```

```python
sns.displot(df_iris['petal_length'], shrink=0.9)
sns.displot(df_iris['petal_length'], kind='kde')
plt.show()
```

```python
sns.displot(df_iris, x='sepal_length', hue='class', col='class', shrink=0.9)
plt.show()
```

```python
sns.displot(df_iris, x='petal_length', hue='class', col='class', shrink=0.9)
plt.show()
```

```python
sns.distplot(df_iris['petal_length'])
plt.show()
```

## Box Plot - Distribution & Comparition

```python
sns.boxplot(data=df_iris)
plt.xlabel('Features')
plt.ylabel('Length in cm')
plt.show()
```

## Pairplot - Relation, Distribution, Comparison

```python
sns.pairplot(df_iris, hue='class')
plt.show()
```

## Heatmap - Correlation / Relation

```python
sns.heatmap(df_iris.corr(), annot=True)
plt.show()
```
