```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn')
```

## Load Data

```python
df_iris = pd.read_csv('./datasets/iris.data', names=['sepal_length', 'sepal_width', 
                                                     'petal_length', 'petal_width', 'class'])
df_iris.head()
```

```python
print('Total rows:', df_iris.shape[0])
print('Total columns:', df_iris.shape[1])
```

```python
df_iris['class'].unique()
```

```python
df_iris.isnull().any()
```

```python
df_iris.describe()
```

## Data Visualization

#### Histogram - Distribution

```python
df_iris['class'].hist()
plt.show()
```

####  Scatter Plot - Distribution & Relation

```python
df_iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
df_iris.plot(kind='scatter', x='petal_length', y='petal_width')
plt.show()
```

#### Distplot - Distribution

```python
sns.distplot(df_iris['sepal_length'])
plt.show()
```

```python
sns.distplot(df_iris['sepal_length'], bins=20)
plt.show()
```

#### Boxplot - Distribution & Comparition

```python
sns.boxplot(data=df_iris)
plt.xlabel('Features')
plt.ylabel('Length in cm')
plt.show()
```

```python
sns.boxplot(data=df_iris, x='class', y='petal_length') # petal_width, sepal_length, sepal_width
plt.xlabel('Class')
plt.ylabel('Length in cm')
plt.show()
```

#### Pairplot - Relation, Distribution, Comparison

```python
sns.pairplot(data=df_iris, hue='class')
plt.show()
```

#### Heatmap - Correlation / Relation

```python
sns.heatmap(data=df_iris.corr(), annot=True)
plt.show()
```
